```python
import re
```

```python
re.sub(r'\.|;|:', '_', 'Hi.....I am;;;;:::robot:::;;;..')
```

```python
re.findall(r'I am \d{1,3}', 'I am 24')
```

```python
re.findall(r'I am (\d{1,3})', 'I am 24')
```

```python
re.findall(r'I am ([^0-1][0-9])', 'I am 21')
```

```python
pattern = re.compile(r'I am ([^0-1][0-9])')
list(filter(pattern.search, ['I am 11', 'I am 21']))
```

```python
# find all phone numbers like `222-34-01`, `2223401`, `222  32  01adas` -> 2223401
```

```python
test_str = 'my number is 222-34-01. my second number is 2223401. Please call me on 222  32  01adas'
re.findall(r'\d{3}-\d{2}-\d{2}|\d{7}|\d+\s+\d+\s+\d+', test_str)
```

```python
# find all phone numbers like `+1222-34-01`, `+1 2223401`, `-1222  32  01adas` -> 2223401
```

```python
test_str = 'my number is +1222-34-01. my second number is 12223401. Please call me on 1222  32  01adas'
re.findall(r'(\+1|1)\s|(\d{3}-\d{2}-\d{2}|\d{7}|\d+\s+\d+\s+\d+)', test_str)
```
